# Parameters Selector | Visuals - Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import noise, random
```

### 100 to 500 points


```python
x = np.linspace(-1,1, 200)
y = [random.random() for i in range(len(x))]
index = 1
for i in range(2):
    for j in range(2):
        plt.subplot(2,2, index)
        plt.plot(x,y)
        plt.title(str(i) + "," + str(j))
        plt.show()
        index+=1
        
```

```python
# 200 points
# varying octaves and persistance
# octaves from 8-10
# persistence from 0.1 - 05
# total plots 5 - 10

octaves_ = [7,8,10]
persistences_ = [0.3, 0.4, 0.5]
size1, size2 = len(octaves_), len(persistences_)
index_ = 1

for i in range(len(octaves_)):
    
    for j in range(len(persistences_)):
        
        x = np.linspace(-1,1,200)
        y = [noise.pnoise1(x[k], octaves=octaves_[i], persistence=persistences_[j]) for k in range(x.size)]
        
        plt.subplot(size1, size2, index_)
        plt.plot(x,y)
        plt.title("Octaves: " + str(octaves_[i]) + " Persistence: " + str(persistences_[j]))
        plt.show()
        
        index_ += 1
```
