# Perlin Noise
So this noise library will finally be understood. Lets see what it has to offer. <br>
https://adrianb.io/2014/08/09/perlinnoise.html <br>
https://github.com/caseman/noise

**octaves**: a noise at a certain amplitude and frequency.<br>
Note: each octave can influence the next one.<br> 
**passes**: bunch of perlin function added together! They have different amplitudes and frequencies. Gives a more naturalistic result. <br>
In other words, Sum of noise functions = (Perlin noise) <br>



```python
import noise
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 0, 200)
y = [noise.pnoise1(x[i], octaves=8, persistence=0.6) for i in range(x.size)]
y = np.array(y)
```

```python
noise.pnoise1(1)
```

```python
plt.plot(x,y)
```
